```python
import urllib.request
import json
import pandas as pd
```

```python
contents = urllib.request.urlopen("https://api.binance.com/api/v1/klines?symbol=LTCBTC&interval=1m")
```

```python
x = contents.read()
```

```python
x
```

```python
y = json.loads(x)
```

```python
df = pd.DataFrame(y,dtype=float)
df.columns=[        'Open_time',
                    'Open',
                    'High',
                    'Low',
                    'Close',
                    'Volume',
                    'Close_time',
                    'Quote_asset_volume',
                    'Number_of_trades',
                    'Taker_buy_base_asset_volume',
                    'Taker_buy_quote_asset_volume',
                    'Ignore']
```

```python
df["Open"].plot()
```
