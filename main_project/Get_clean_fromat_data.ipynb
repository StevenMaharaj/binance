```julia
using HTTP
using JSON
using DataFrames
```

```julia
url = "https://api.binance.com/api/v1/time";
res = HTTP.get(url);
```

```julia
function get_data(symbol = "BTCUSDT",interval = "3m")
    response = HTTP.get("https://api.binance.com/api/v1/klines?symbol="*symbol*"&interval="*interval)
#     response = HTTP.get("https://api.binance.com/api/v1/klines?symbol=BTCUSDT&interval=3m")
    data = JSON.parse(String(response.body))
end
    
```

```julia
data = get_data()
d =reshape(data[1],(1,12))
df = DataFrame(d)
for i in 2:length(data)
    push!(df,reshape(data[i],(1,12)))
end
rename!(df,:x1=>:Open_time,
    :x2=>:Open,
    :x3=>:High,
    :x4=>:Low,
    :x5=>:Close,
    :x6=>:Volume,
    :x7=>:Close_time,
    :x8=>:Quote_asset_volume,
    :x9=>:Number_of_trades,
    :x10=>:Taker_buy_base_asset_volume,
    :x11=>:Taker_buy_quote_asset_volume,
    :x12=>:Ignore)
```
